INTRODUCTION

Analysing retail transaction data provide valuable insights into customer behavior,purchasing patterns and sales trends.This information can be used to inform business decisions,optimize marketing strategies and improve overall customer satisfaction.

IMPORT LIBRARIES

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
data=pd.read_csv(r"D:\data_analytics\project\RETAIL_TRANSACTION_DATA.csv")
data=data.drop('Zipcode',axis=1)
data.info()
data.head(10)
```

DATA CLEANING

```python
sns.boxplot(data=data)
plt.xticks(rotation=45)
plt.show()
```

```python
data.isna().sum()
data.dropna(inplace=True)

```

```python
data['Date']=pd.to_datetime(data['Date'],errors='coerce')
data['Time']=pd.to_datetime(data['Time'],errors='coerce')
data.info()
```

AVERAGE PURCHASE AMOUNT

```python

average_purchase_amont=data['Total_Amount'].mean()
print(average_purchase_amont)
```

TOTAL SALES TAKES PLACE IN EVERY MONTH

```python

seasonal_selling_product=data.groupby('Month')['Product_Category'].size().reset_index(name='count')
print(seasonal_selling_product.head(12))
plt.figure(figsize=(6, 6))
sns.barplot(x='Month', y='count', data=seasonal_selling_product, palette='viridis')
plt.xlabel("Month")
plt.ylabel("Number of Products Sold")
plt.title("Monthly Sales Distribution", fontsize=14)
plt.xticks(rotation=45)  
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

SEASONAL TRENDS IN SALES

```python
def get_season(month):
    if month in('December','January','February'):
        return 'winter'
    elif month in('March','April','May'):
        return 'summer'
    elif month in ('June','July','August','September'):
        return 'mansoon'
    else:
        return 'autum'
data['season']=data['Month'].apply(get_season)
seasonal_trend = data.groupby(['season', 'Product_Category'])['Product_Category'].size().unstack()
print(seasonal_trend)
plt.figure(figsize=(8, 8))
seasonal_trend.plot(kind='bar', stacked=False, colormap='viridis', figsize=(6, 6))
plt.xlabel("Season")
plt.ylabel("Count of Products Sold")
plt.title("Seasonal Trend of Product Categories")
plt.xticks(rotation=45)
plt.legend(title="Product Category")
plt.show()

```

 CATEGORY OF CUSTOMERS WITH THEIR BUYING POWER

```python
type_of_customers=data.groupby('Customer_Segment').size().sort_values(ascending=False).reset_index(name='count')
print(type_of_customers)
plt.figure(figsize=(6,4))
plt.pie(type_of_customers['count'], labels=type_of_customers['Customer_Segment'], autopct='%1.1f%%')
plt.title("Customer Segment Distribution")
plt.show()
```

MOST POPULAR PAYMENT METHOD

```python
popular_payment_method=data.groupby("Payment_Method").size().sort_values(ascending=False).reset_index(name='count')
print(popular_payment_method)
plt.figure(figsize=(6,4))
plt.plot(popular_payment_method['Payment_Method'],popular_payment_method['count'],marker='o',linestyle='-',color='b')
plt.xlabel('Payment_Method')
plt.ylabel('count')
plt.title('most polular payment method')
plt.xticks(rotation=45)
plt.show()
```

PRODUCT BRAND THAT MOSTLY PURCHASED BY CUSTOMERS

```python
popular_brand=data.groupby('Product_Brand').size().reset_index(name='count').sort_values(by='count',ascending=False)
print(popular_brand)
plt.figure(figsize=(6, 6))
plt.pie(popular_brand['count'], labels=popular_brand['Product_Brand'], autopct='%1.1f%%')
plt.title("Market Share of Product Brands")
plt.show()


```

THE AGE DISTRIBUTION OF CUSTOMERS MAKING PURCHASE

```python
age_bins = [0, 18, 25, 35, 45, 55, 65, 100]
age_labels = ["0-18", "19-25", "26-35", "36-45", "46-55", "56-65", "65+"]
data["age_group"] = pd.cut(data["Age"], bins=age_bins, labels=age_labels)
age_distribution = data["age_group"].value_counts().sort_index()
print(age_distribution)
plt.figure(figsize=(6,6))
sns.barplot(x=age_distribution.index, y=age_distribution.values, palette="viridis")
plt.xlabel("Age Group")
plt.ylabel("Number of Customers")
plt.title("Customer Distribution by Age Group")
plt.xticks(rotation=45)
plt.show()
```

```python
age_group_with_customer_type=data[['age_group','Customer_Segment']].value_counts().reset_index(name='count')
print(age_group_with_customer_type)
```

REPEAT PURCHASE RATE OF CUSTOMERS

```python
repeat_purchase=data.groupby("Customer_ID")['Total_Purchases'].count()
repeat_cutomer=repeat_purchase[repeat_purchase > 1].count()
total_customers=repeat_purchase.count()
repeat_purchase_rate=(repeat_cutomer/total_customers)*100
print(repeat_purchase_rate)
print(f"Repeat Purchase Rate: {repeat_purchase_rate:.2f}%")
```

GENDER IMPACT ON PRODUCT CATEGORY

```python
gender_impact=data.groupby(['Gender','Product_Category'])['Product_Category'].size()
print(gender_impact)
plt.figure(figsize=(6, 6))
sns.countplot(data=data, x='Product_Category', hue='Gender', order=gender_impact.index.get_level_values(1).unique(),palette='dark')
plt.title('Product Category Distribution by Gender')
plt.xlabel('Product Category')
plt.ylabel('Count of Products')
plt.xticks(rotation=45, ha="right")
plt.tight_layout()
plt.show()


```

PRODUCT THAT GENERATE HIGHEST REVENUE


```python

revenue=data.groupby('Product_Category')['Total_Amount'].sum().round(2).sort_values(ascending=False)
print(revenue)
plt.figure(figsize=(6,6))
plt.pie(revenue, labels=revenue.index, autopct='%1.1f%%', colors=sns.color_palette("viridis", len(revenue)))
plt.title("Revenue Distribution by Product Category")
plt.show()
```

CUSTOMERS PREFFERED SHIPPING METHOD

```python
shipping_method=data.groupby('Shipping_Method').size().sort_values(ascending=False)
print(shipping_method)
shipping_method = shipping_method.reset_index()
shipping_method.columns = ['Shipping_Method', 'Count']
plt.figure(figsize=(5, 5))
sns.barplot(x='Shipping_Method', y='Count', data=shipping_method,palette='dark')
plt.title("Distribution of Shipping Methods")
plt.xticks(rotation=45)
plt.show()

```

TOP SELLING PRODUCT

```python
product_with_highest_sale=data.groupby('Product_Category')['Total_Purchases'].size().sort_values(ascending=False).reset_index(name='count')
print(product_with_highest_sale)
plt.figure(figsize=(8, 4))
sns.barplot(y='Product_Category', x='count', data=product_with_highest_sale, palette='muted')
plt.xlabel("Number of Purchases")
plt.ylabel("Product Category")
plt.title("Top Selling Product Categories")
plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.xticks(rotation=45)  
plt.show()
```

CORRELATION BETWEEN RATING,TOTAL PURCHASE AND PRICE OF THE PRODUCT

```python
rating_product=data.groupby('Product_Category')['Ratings'].size().sort_values(ascending=False)
corr_matrix=data[['Ratings','Amount','Total_Purchases']].corr()
print(rating_product)
print(corr_matrix)
sns.heatmap(corr_matrix,annot=True)
plt.title("relation between price and rating")
plt.show()

```

CITTY WITH HIGHEST SALES TAKES PLACE

```python
citty_with_sales=data.groupby('City')['Total_Purchases'].size().sort_values(ascending=False).reset_index(name='count')
highest_sales_takesplace=citty_with_sales.nlargest(10,'count')
print('city with highest sales')
print(highest_sales_takesplace)
plt.figure(figsize=(8,4))
sns.barplot(x='count', y='City', data=highest_sales_takesplace, palette='viridis')
plt.xlabel("Number of Purchases")
plt.ylabel("City")
plt.title("Top 10 Cities with Highest Sales", fontsize=14)
plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.show()


```
